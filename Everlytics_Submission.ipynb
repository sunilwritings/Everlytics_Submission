```python
df1 = input1[['ItemId','ItemName','Comment']]
```

```python
df1.head(1)
```

```python
df2=input2.drop(['Milli Sec'],axis=1)
```

```python
df2.head(5)
```

```python
df2.insert(2, "Time_Stamp", df2['Date'].astype(str) +" "+ df2["Time"].astype(str), True)
```

```python
df2=df2.drop(columns=['Date','Time'],axis=1)
```

```python
df2.head(2)
```

```python
df2['Time_Stamp'] = pd.to_datetime(df2['Time_Stamp'], errors='coerce')
```

```python
df5=df2.groupby(pd.Grouper(key='Time_Stamp', freq='10s')).max()
```

```python
df5.head(1)
```

```python
Tag_values=[]
Time_Stamp=[]
for i in df1['ItemName']:
  for j in df5.columns:
    if i==j:
      Tag_values.append(max(df5[i]))
      Time_Stamp.append(df5[j])
```

```python
df1['Tag_values']=Tag_values
df1['Event_Time_stamp']=Time_Stamp
```

```python
df1['Tag_ID']=df1['ItemId']
df1['Tag_Name']=df1['ItemName']
df1['tag__desc']=df1['Comment']
```

```python
df1.head(1)
```

```python
df1=df1.drop(columns=['ItemId','ItemName','Comment'])
```

```python
df1['Unit']=df1['tag__desc'].str.extract(r'\((\w+)\)', expand=True)
```

```python
df1.head(1)
```

```python
df1=df1[['Event_Time_stamp','Tag_ID','Tag_Name','tag__desc','Tag_values','Unit']]
```
